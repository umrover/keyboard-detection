```python
import glob
import random

import matplotlib.pyplot as plt
import numpy as np

from PIL import Image
from keyrover import TEST_DATASET, imhist
```

```python
files = glob.glob(f"{TEST_DATASET}/*")
```

```python
img = Image.open(random.choice(files))
img
```

```python
kwargs={"bins": 50, "alpha": 0.5, "histtype": "stepfilled"}
imhist(img, **kwargs)
```

```python
n = int(np.ceil(len(files) ** 0.5))
_, axes = plt.subplots(nrows=n, ncols=n)

for i, path in enumerate(files):
    img = Image.open(path)
    imhist(img, axes[i // n][i % n], **kwargs)

for i in range(n ** 2):
    axes[i // n][i % n].axis("off")
```

```python
# Average Histogram

images = np.array([Image.open(path) for path in files])
imhist(images, **kwargs)
```
