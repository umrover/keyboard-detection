```python
import pickle

from keyrover.datasets import id_to_key
from keyrover.color import Palette
from keyrover import *
```

```python
with open(f"{RAW_TEXCOORDS}/key_texcoords_means.bin", "rb") as file:
    U_means = pickle.load(file)
    V_means = pickle.load(file)

with open(f"{RAW_TEXCOORDS}/key_texcoords_all.bin", "rb") as file:
    U = pickle.load(file)
    V = pickle.load(file)


ax = plt.gca()
ax.hist2d(U, V, bins=75)

for (cls, u), (_, v) in zip(U_means.items(), V_means.items()):
    ax.text(u, v, id_to_key[cls], c="white", fontsize=8)

plt.show()
```

```python
colors = []
classes = []

for (cls, u), (_, v) in zip(U_means.items(), V_means.items()):
    colors.append((u, v))
    classes.append(cls)

palette = Palette(colors, classes)
```
